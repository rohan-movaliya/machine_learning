```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer_dataset = load_breast_cancer()
```

```python
breat_cancer_data = pd.DataFrame(breast_cancer_dataset.data, columns=breast_cancer_dataset.feature_names)
```

```python
breat_cancer_data["label"] = breast_cancer_dataset.target
```

```python
X = breat_cancer_data.drop(columns="label", axis=1)
Y = breat_cancer_data["label"]
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
X_train = X_train_std
X_test = X_test_std
```

```python
print(X_train)
```

```python
print(X_test)
```

#**Building the Neural Network**

download.png

```python
# importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layer of Neural Network
model = keras.Sequential([
                            keras.layers.Flatten(input_shape=(30,)),
                            keras.layers.Dense(20, activation="relu"),
                            keras.layers.Dense(2, activation="sigmoid")
])
```

```python
# Compiling the Neural Network
model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics = ["accuracy"])
```

```python
# Training the Neural Network
history = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

#**Visulizing Accuracy & Loss**

```python
plt.plot(history.history["accuracy"])
plt.plot(history.history["val_accuracy"])

plt.title("Model Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")

plt.legend(["Training Data", "Validation Data"], loc="lower right")
```

```python
plt.plot(history.history["loss"])
plt.plot(history.history["val_loss"])

plt.title("Model Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")

plt.legend(["Training Data", "Validation Data"], loc="upper right")
```

```python
# Accuracy of the model on training data
loss,accuracy = model.evaluate(X_test, Y_test)
print(accuracy)
```
